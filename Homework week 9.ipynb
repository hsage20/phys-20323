```python
from random import random
import numpy as np

n_simulations = 1000
# a fluffy teddybear is the prize
noswitch_results = []
switch_results = []
for i in range(n_simulations):
    doors = ['fluffy teddybear', 'nada', 'nada']
    np.random.choice(doors)
    my_door_index = np.random.choice([0, 1, 2])
    noswitch_result = doors.pop(my_door_index)
    doors.remove('nada')
    switch_result = doors[0]
    noswitch_results.append(noswitch_result)
    switch_results.append(switch_result)

noswitch_results.count('fluffy teddybear') / n_simulations
switch_results.count('nada')/ n_simulations
```
